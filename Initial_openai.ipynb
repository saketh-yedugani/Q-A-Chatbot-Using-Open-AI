```python
import os
from dotenv import load_dotenv
load_dotenv()


os.environ['OPENAI_API_KEY']=os.getenv("OPENAI_API_KEY")
## Langsmith Tracking
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]="true"
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
```

```python
from langchain_openai import ChatOpenAI
llm=ChatOpenAI(model="gpt-4o")  #We are using gpt-4o model here
print(llm)
```

```python
## Input and get response form LLM

result=llm.invoke("What are Northern Lights?")
print(result)
```

```python
# Using Chatprompt Template we can give in a input prompt to the LLM model so that it can give response based on that prompt.
from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate.from_messages(
    [
        ("system","You are an expert Geologist. Provide me answers based on the questions"),
        ("user","{input}")
    ]

)
prompt
```

```python
## chain up the prompt and llm
chain=prompt|llm        #input->prompt->llm->output

response=chain.invoke({"input":"Can you tell me about Northeren Lights?"})
print(response)
```

```python
## Using stroutput Parser we also can parse the output from the LLM model as string

from langchain_core.output_parsers import StrOutputParser
output_parser=StrOutputParser()
chain=prompt|llm|output_parser      #input->prompt->llm->output_parser->output

response=chain.invoke({"input":"Can you tell me about Northeren Lights?"})
print(response)
```
